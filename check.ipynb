```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import display
from ipywidgets import FileUpload

# Load your trained model
model = tf.keras.models.load_model('model10.h5')

# Create a file upload widget
uploader = FileUpload(accept='.jpeg', multiple=False)

# Define a function to handle the file upload and make predictions
def handle_upload(change):
    uploaded_filename = list(uploader.value.keys())[0]
    uploaded_data = uploader.value[uploaded_filename]['content']
    
    # Create a temporary file to save the uploaded image
    with open(uploaded_filename, 'wb') as f:
        f.write(uploaded_data)
    
    # Load the uploaded image
    img = Image.open(uploaded_filename)
    img = np.array(img) / 255.0  # Normalize the image to [0, 1]

    # Make predictions on the uploaded image
    x = np.expand_dims(img, axis=0)
    predictions = model.predict(x)
    predicted_class = 'DRONES' if predictions[0] > 0.5 else 'BIRDS'

    # Display the uploaded image and prediction
    plt.imshow(img)
    plt.title(f'Prediction: {predicted_class}')
    plt.axis('off')
    plt.show()

uploader.observe(handle_upload, names='value')
display(uploader)
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image

# Load your trained model
model = tf.keras.models.load_model('model10.h5')

# List of image paths
image_paths = ['bird1.jpeg', 'bird2.jpeg', 'bird3.jpeg', 'bird4.jpeg', 'bird.jpeg', 'bird.jpg', 'drone.jpg', 'drone1.jpeg', 'drone2.jpeg', 'drone3.jpeg', 'drone4.jpeg','drone5.jpeg']

# Calculate the number of rows and columns for subplots
num_images = len(image_paths)
num_rows = (num_images // 5) + (1 if num_images % 5 > 0 else 0)
num_cols = min(num_images, 5)

# Set up subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 6))

for i, image_path in enumerate(image_paths):
    img = Image.open(image_path)  # Load the actual image
    img = img.resize((28, 28))  # Resize the image to (28, 28)
    img = np.array(img) / 255.0  # Normalize the image to [0, 1]

    # Make predictions on the resized image
    x = np.expand_dims(img, axis=0)
    predictions = model.predict(x)
    predicted_class = 'DRONE' if predictions[0] > 0.5 else 'BIRD'

    # Calculate row and column for the current image
    row = i // 5
    col = i % 5

    # If there's only one row, the axes array will be one-dimensional
    if num_rows == 1:
        axes[col].imshow(img)
        axes[col].set_title(f'File: {os.path.basename(image_path)}\nPrediction: {predicted_class}')
        axes[col].axis('off')
    else:
        axes[row, col].imshow(img)
        axes[row, col].set_title(f'File: {os.path.basename(image_path)}\nPrediction: {predicted_class}')
        axes[row, col].axis('off')

# Remove any empty subplots
for i in range(num_images, num_rows * num_cols):
    if num_rows == 1:
        fig.delaxes(axes[i])
    else:
        row = i // 5
        col = i % 5
        fig.delaxes(axes[row, col])

plt.tight_layout()
plt.show()
```
